```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import os
train=pd.read_csv('/kaggle/input/playground-series-s3e16/train.csv',index_col=False)
train_age=train['Age']
train.pop('Age')
train.pop('id')
le=preprocessing.LabelEncoder()
label = le.fit_transform(train['Sex'])
train.drop('Sex', axis=1, inplace=True)
train['Sex'] = label
train=np.asarray(train).astype(np.float32)
train_age=np.asarray(train_age).astype(np.float32)
```

```python
test=pd.read_csv('/kaggle/input/playground-series-s3e16/test.csv',index_col=False)
test.pop('id')
le=preprocessing.LabelEncoder()
label = le.fit_transform(test['Sex'])
test.drop('Sex', axis=1, inplace=True)
test['Sex'] = label
test=np.asarray(test).astype(np.float32)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train = scaler.fit_transform(train)
test = scaler.fit_transform(test)
model = LogisticRegression(random_state=2,solver='lbfgs', max_iter=10000,penalty='l2').fit(train,train_age)
arr=model.predict(test)
print(arr)
```

```python
import csv
test=pd.read_csv('/kaggle/input/playground-series-s3e16/test.csv',index_col=False)
file="submission.csv"
with open(file, 'w') as csvfile: 
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerow(['id','Age'])
    for id1 in test['id']:
        csvwriter.writerows([[id1]])
arr=[int(i) for i in arr]
filecsv=pd.read_csv(file)
filecsv["Age"]=arr
filecsv.to_csv(file,index=False)
print(filecsv)
print(pd.read_csv('/kaggle/working/submission.csv'))
```
